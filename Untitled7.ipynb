```python
import os
import pandas as pd

# Path to the audio files (adjust if the folder name is different)
data_path = '/content/audio_speech_actors_01-24'

# Collect all .wav files
file_paths = []
emotions = []

for actor_folder in os.listdir(data_path):
    actor_path = os.path.join(data_path, actor_folder)
    if os.path.isdir(actor_path):
        for file in os.listdir(actor_path):
            if file.endswith('.wav'):
                file_paths.append(os.path.join(actor_path, file))
                # Extract emotion from filename: e.g., 03-01-06-01-02-01-12.wav → emotion code is index 2
                parts = file.split('-')
                emotion_code = int(parts[2])
                emotions.append(emotion_code)

# Map codes to emotion names
emotion_map = {
    1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad',
    5: 'angry', 6: 'fearful', 7: 'disgust', 8: 'surprised'
}
emotion_labels = [emotion_map.get(code, 'unknown') for code in emotions]

df = pd.DataFrame({'path': file_paths, 'emotion': emotion_labels})
print("✅ Total files loaded:", len(df))
print("\nFirst 5 rows:")
print(df.head())
print("\nEmotion distribution:")
print(df['emotion'].value_counts())
```

```python
import librosa
import numpy as np
from tqdm import tqdm

def extract_mfcc(file_path):
    try:
        audio, sr = librosa.load(file_path, duration=3, sr=22050, offset=0.5)
        mfcc = np.mean(librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40).T, axis=0)
        return mfcc
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

print("Extracting features... This may take a few minutes.")

features = []
labels = []

for idx, row in tqdm(df.iterrows(), total=len(df)):
    feat = extract_mfcc(row['path'])
    if feat is not None:
        features.append(feat)
        labels.append(row['emotion'])

X = np.array(features)
y = np.array(labels)

print("\n✅ Feature extraction done!")
print("Feature shape (samples, mfcc_features):", X.shape)
print("Labels shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded)

# Train model
model = MLPClassifier(hidden_layer_sizes=(256, 128), max_iter=300, alpha=0.01, random_state=42)
model.fit(X_train, y_train)

# Predict & Evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("✅ Training Complete!")
print("Test Accuracy:", round(accuracy * 100, 2), "%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import pandas as pd
import os

# Load or create metadata
data_path = '/content/audio_speech_actors_01-24'  # adjust if needed

file_paths = []
emotions = []

for actor in os.listdir(data_path):
    actor_path = os.path.join(data_path, actor)
    if os.path.isdir(actor_path):
        for file in os.listdir(actor_path):
            if file.endswith('.wav'):
                file_paths.append(os.path.join(actor_path, file))
                parts = file.split('-')
                emotion_code = int(parts[2])
                emotions.append(emotion_code)

emotion_map = {1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad', 5: 'angry', 6: 'fearful', 7: 'disgust', 8: 'surprised'}
df = pd.DataFrame({'path': file_paths, 'emotion': [emotion_map[c] for c in emotions]})

print("Total audio files:", len(df))
print("\nEmotion distribution:\n", df['emotion'].value_counts())

# Save metadata
df.to_csv('data/metadata.csv', index=False)
print("\n✅ Metadata saved to data/metadata.csv")
```

```python
import numpy as np
from tqdm import tqdm

def extract_mfcc(file_path, n_mfcc=40):
    """Extract mean MFCC features from audio file."""
    try:
        audio, sr = librosa.load(file_path, duration=3, sr=22050, offset=0.5)
        mfcc = np.mean(librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc).T, axis=0)
        return mfcc
    except:
        return None

print("Extracting MFCC features from all files...")

features = []
labels = []

for idx in tqdm(range(len(df))):
    path = df['path'].iloc[idx]
    emotion = df['emotion'].iloc[idx]
    feat = extract_mfcc(path)
    if feat is not None:
        features.append(feat)
        labels.append(emotion)

X = np.array(features)
y = np.array(labels)

np.save('data/features.npy', X)
np.save('data/labels.npy', y)

print("\n✅ Feature extraction complete!")
print("Features shape:", X.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import joblib

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded)

model = MLPClassifier(hidden_layer_sizes=(256, 128), max_iter=300, alpha=0.01, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Test Accuracy:", round(acc*100, 2), "%")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Save
joblib.dump(model, 'models/ser_mlp.pkl')
joblib.dump(le, 'models/label_encoder.pkl')
print("Model saved!")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.savefig('reports/confusion_matrix.png')
plt.show()
```
